```
import ipywidgets as widgets
from IPython.display import display

# สร้างข้อความต้อนรับ
title = widgets.HTML(
    value="<h2>💖  ถึงแฟนที่น่ารัก  💖</h2>",
    placeholder='Some HTML',
    description=''
)

# สร้างกล่องใส่ชื่อ
name_input = widgets.Text(
    placeholder='ใส่ชื่อนะครับเบบี้😘.',
    description='ถึง:',
    disabled=False
)

# สร้างปุ่มกด
button = widgets.Button(
    description='กดตรงเลย ! 💕',
    button_style='danger', # 'success', 'info', 'warning', 'danger' หรือ ''
    tooltip='Click me',
    icon='heart'
)

# สร้างพื้นที่แสดงผลลัพธ์
output = widgets.Output()

# ฟังก์ชันแสดงคำบอกรัก
def on_button_clicked(b):
    with output:
        output.clear_output()
        if name_input.value == "":
            print("💖 ถึง เชียร์แฟนเค้า อยู่กับเค้าไปนานๆ ตั้งใจเรียน กินข้าวให้ตรงเวลา อย่านอนดึกมาก เค้าเป้นห่วง และขอบคุณที่อยู่ด้วยกันมาจนถึงตอนนี้เค้ามีความสุขมากๆเลยนะรักนะจุ๊บๆ! 😘 ")
        else:
            print(f"💖 ถึง {name_input.value}  อยู่กับเค้าไปนานๆ ตั้งใจเรียน กินข้าวให้ตรงเวลา อย่านอนดึกมาก เค้าเป้นห่วงอยู่กับเค้าไปนานๆ ตั้งใจเรียน กินข้าวให้ตรงเวลา อย่านอนดึกมาก เค้าเป้นห่วงรักนะจุ๊บๆ! 😘รักนะจุ๊บๆ! 😘")

# เชื่อมปุ่มเข้ากับฟังก์ชัน
button.on_click(on_button_clicked)

# แสดงผลทั้งหมดบน Colab
display(title, name_input, button, output)
```
